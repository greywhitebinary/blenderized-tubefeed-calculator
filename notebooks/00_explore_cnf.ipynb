## 3. Spot-check: look up a food and its nutrients

Let's find "Chicken, breast" and see its protein value.

```python
food = tables["food_name"]
chicken = food[food["Food_Description_EN"].str.contains("chicken", case=False, na=False)]
chicken.head(10)
```

```python
# Pick a Food_Code and look up its nutrient amounts
code = chicken.index[0]  # first match
print(f"Looking up Food_Code = {code}")

amounts = tables["nutrient_amount"]
nutrients = tables["nutrient_name"]

merged = (
    amounts[amounts["Food_Code"] == code]
    .merge(nutrients, on="Nutrient_Code")
)
merged[["Nutrient_Name_EN", "Nutrient_Amount", "Nutrient_Unit"]].head(20)
```
